```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
###load the dataset
df= pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
###removing unwanted columns
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
#encoding the categorical variable
label_encode_gender=LabelEncoder()
df['Gender']=label_encode_gender.fit_transform(df['Gender'])
df
#converted female as 0 and male as 1
```

```python
#one hot encoding on feature Geography
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder_geography=OneHotEncoder()
one_hot_geography=one_hot_encoder_geography.fit_transform(df[['Geography']])
one_hot_geography
```

```python
one_hot_encoder_geography.get_feature_names_out(['Geography'])
```

```python
one_hot_geography.toarray()
```

```python
geo_encoded_df=pd.DataFrame(one_hot_geography.toarray(),columns=one_hot_encoder_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
#combining the  one hot encoded column  with original dataframe
data= pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
data
```

```python
#saving the one hot encoded data and scalar file in pkl format
with open('label_encode_gender.pkl', 'wb') as file:
    pickle.dump(label_encode_gender,file)

with open('one_hot_encoder_geography.pkl','wb') as file:
    pickle.dump(one_hot_encoder_geography,file)
    
    
```

```python
#Dividing the data into dependent and independent set
y=data['Exited']
x=data.drop('Exited', axis=1)

x_train,x_test, y_train,y_test= train_test_split(x,y, test_size=0.2, random_state=42)


#sCALLING DOWN THE FEATRURE
scalling=StandardScaler()
x_train=scalling.fit_transform(x_train)
x_test=scalling.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#saving the scaling file into the form of pickle file
with open('scalling.pkl','wb') as file:
    pickle.dump(scalling,file)
   
```

```python
#Ann Implementation
import tensorflow as tf
from tensorflow.python.keras.layers import Dense
from tensorflow import keras
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Sequential, load_model
from tensorflow.python.keras.callbacks import TensorBoard,EarlyStopping
import datetime
```

```python
(x_train.shape[1],)
```

```python
#building a model
model=Sequential([
Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
Dense(32,activation='relu'),
Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt= tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
#setting up the tenserboard
import datetime
from tensorflow.python.keras.callbacks import TensorBoard,EarlyStopping
logs_dir="logs/fit"+ datetime.datetime.now().strftime("%Y%M%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=logs_dir,histogram_freq=1)
```

```python
#setting early stopping
early_stopping_callbacks=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
#model trainning
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50,callbacks=[early_stopping_callbacks])
```

```python
model.save('model.h5')
```
